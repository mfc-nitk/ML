```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
print(df.head())
plt.scatter(df['x1'].head(200), df['y'].head(200))
```

```python
m, n = df.shape
n = n-1
x = np.zeros((m, n+1))
x[:, 0] += np.ones(m)
x[:, 1:] += np.array(df)[:, :-1]

y = np.array(df['y'])
y.resize(m ,1)

# print(x.shape)
# print(x)
# print(y.shape)
# print(y)
```

```python
import random

train_size = int(0.75*m)

train_indices = random.sample(range(m), train_size)
train_indices = np.array(train_indices)

test_indices = []
for i in range(m):
    if i not in train_indices:
        test_indices.append(i)

test_indices = np.array(test_indices)


x_train = x[train_indices, :]
y_train = y[train_indices]

x_test = x[test_indices, :]
y_test = y[test_indices]

# print(train_indices)
# print(x_train)
# print(y_train)

# print(test_indices)
# print(x_test)
# print(y_test)

# print('{} + {} = {}'.format(train_indices.shape[0], test_indices.shape[0], train_indices.shape[0]+test_indices.shape[0]))
# print(m)
```

```python
def cost(x, y, theta):
    return 0.5*((np.matmul(x, theta) - y)**2).sum()

def gradient_descent(x, y, theta, alpha, iterations):
    
    for i in range(iterations):
        theta = theta - alpha*np.matmul(x.transpose(), (np.matmul(x, theta)-y))
        cst = cost(x, y, theta)
        print('iteration:{}, cost:{}, theta:{}'.format(i, cst, theta.reshape(1, theta.shape[0]*theta.shape[1])))
    return theta
```

```python
alpha = 0.00005

theta = np.zeros((n+1,1))
print(theta)

theta = gradient_descent(x_train, y_train, theta, alpha, 500)

print(theta)
```

```python
y_predict = np.matmul(x_test, theta)

plt.plot(x_test[:, 1], y_predict, 'k')
plt.scatter(x[:, 1], y)
```
